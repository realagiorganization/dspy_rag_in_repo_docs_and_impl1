## 1. Load The Modular Training Helpers

### 1.1 Keep Training Preparation In Python Modules

Notebook cells should orchestrate the workflow, not reimplement normalization and validation rules inline.


```python
from pathlib import Path

from repo_rag_lab.dspy_training import resolve_dspy_lm_config
from repo_rag_lab.dspy_workflow import RepositoryRAG
from repo_rag_lab.notebook_scaffolding import build_training_lab_context
from repo_rag_lab.notebook_support import (
    assert_minimum_pass_rate,
    assert_no_validation_issues,
    resolve_repo_root,
    write_notebook_run_log,
)
from repo_rag_lab.training_samples import batch_training_examples, load_training_examples

root = resolve_repo_root(Path.cwd().resolve())
training_path = root / "samples" / "training" / "repository_training_examples.yaml"
examples = load_training_examples(training_path)
context = build_training_lab_context(root)
len(examples)
```

### 3.1 Prepare Small Review Units Before DSPy Optimization

```python
batch_training_examples(examples, batch_size=2)
```

## 4. Inspect Or Reuse The Latest Compiled Program

### 4.1 Load The Saved Artifact When LM Configuration Is Available

The training-lab scaffold exposes the latest compiled DSPy artifact metadata. If a saved program exists and LM configuration is available in this process, reuse it against a training question; otherwise inspect the artifact metadata and the next step.


```python
lm_config = resolve_dspy_lm_config()
if context["compiled_program_path"] and lm_config is not None:
    compiled_runtime = RepositoryRAG(
        root,
        program_path=root / context["compiled_program_path"],
        lm_config=lm_config,
        require_configured_lm=True,
    )
    compiled_result = compiled_runtime(examples[0].question)
    {
        "compiled_program_path": context["compiled_program_path"],
        "compiled_program_summary": context["compiled_program_summary"],
        "question": examples[0].question,
        "answer": compiled_result.answer,
        "context_preview": compiled_result.context[:1],
    }
else:
    {
        "compiled_program_path": context["compiled_program_path"],
        "compiled_program_metadata_path": context["compiled_program_metadata_path"],
        "next_step": (
            "Run `make dspy-train ...` and provide LM configuration to reuse the saved "
            "program here."
        ),
    }
```

## 5. Validate Assertions And Log The Run

### 5.1 Record Benchmark Health And Tuning Metadata

Use the scaffold output to assert benchmark quality and persist both tuning metadata and notebook logs.


```python
assert_no_validation_issues(context["validation_issues"], context="training samples")
assert_minimum_pass_rate(context["benchmark_summary"], minimum_pass_rate=1.0)
log_path = write_notebook_run_log(root, "03-dspy-training-lab", context)
{
    "tuning_metadata_path": context["tuning_metadata_path"],
    "log_path": str(log_path.relative_to(root)),
}
```
